Añadimos una columna por cada característica al dataframe obteniendo un dataframe final con todas las características calculadas

```python
# Añade una nueva columna con valores 0 o 1 dependiendo de si la columna 'Category' es 'Ham' o 'Spam'
df['Class'] = df['Category'].apply(lambda x: 0 if x == 'ham' else 1)

def calcular_frecuencia_palabra(mensaje, palabra):
    # Divide el mensaje en palabras
    palabras = mensaje.split()
    # Manejar el caso donde no hay palabras en el mensaje
    if len(palabras) == 0:
        return 0
    # Calcula la frecuencia relativa de la palabra
    frecuencia_palabra = 100 * palabras.count(palabra) / len(palabras)
    return frecuencia_palabra

def calcular_frecuencia_caracter(mensaje, caracter):
    # Manejar el caso donde el mensaje está vacío
    if len(mensaje) == 0:
        return 0
    # Calcula la frecuencia relativa del carácter
    frecuencia_caracter = 100 * mensaje.count(caracter) / len(mensaje)
    return frecuencia_caracter

def longitud_promedio_de_secuencias(mensaje):
    # Lista para almacenar las longitudes de las secuencias ininterrumpidas de letras mayúsculas
    longitudes_secuencias = []

    # Variable para almacenar la longitud de la secuencia actual
    longitud_actual = 0
    for caracter in mensaje:
        if caracter.isupper():
            # Si encontramos una letra mayúscula, aumentamos la longitud de la secuencia actual
            longitud_actual += 1
        else:
            # Si no es una letra mayúscula, almacenamos la longitud de la secuencia actual
            if longitud_actual > 0:
                longitudes_secuencias.append(longitud_actual)
                longitud_actual = 0  # Reiniciamos la longitud de la secuencia actual

    # Si el mensaje termina con una secuencia ininterrumpida de letras mayúsculas, la almacenamos
    if longitud_actual > 0:
        longitudes_secuencias.append(longitud_actual)

    # Calculamos el promedio de las longitudes de las secuencias ininterrumpidas de letras mayúsculas
    if longitudes_secuencias:
        longitud_promedio = sum(longitudes_secuencias) / len(longitudes_secuencias)
        return longitud_promedio
    else:
        return 0  # Si no hay secuencias de letras mayúsculas, el promedio es cero

def longitud_de_la_secuencia_mas_larga(mensaje):
    longitud_maxima = 0
    longitud_actual = 0

    for caracter in mensaje:
        if caracter.isupper():
            # Si encontramos una letra mayúscula,
            # aumentamos la longitud de la secuencia actual
            longitud_actual += 1
        else:
            # Si no es una letra mayúscula, almacenamos la longitud
            # de la secuencia actual si es mayor que la longitud máxima
            if longitud_actual > longitud_maxima:
                longitud_maxima = longitud_actual
            # Reiniciamos la longitud de la secuencia actual
            longitud_actual = 0

    # Verificamos si la última secuencia es la más larga
    if longitud_actual > longitud_maxima:
        longitud_maxima = longitud_actual

    return longitud_maxima

def numero_total_de_letras_mayusculas(mensaje):
    total_mayusculas = sum(1 for char in mensaje if char.isupper())
    return total_mayusculas

print(calcular_frecuencia_palabra('FreeMsg Hey there darling it''s been 3 week''s now and no word back! I''d like some fun you up for it still? Tb ok! XxX std chgs to send, £1.50 to rcv','you'))


df['word_freq_make'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='make')
df['word_freq_address'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='address')
df['word_freq_all'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='all')
df['word_freq_3d'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='3d')
df['word_freq_our'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='our')
df['word_freq_over'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='over')
df['word_freq_remove'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='remove')
df['word_freq_internet'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='internet')
df['word_freq_order'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='order')
df['word_freq_mail'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='mail')
df['word_freq_receive'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='receive')
df['word_freq_will'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='will')
df['word_freq_people'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='people')
df['word_freq_report'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='report')
df['word_freq_addresses'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='addresses')
df['word_freq_free'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='free')
df['word_freq_business'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='business')
df['word_freq_email'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='email')
df['word_freq_you'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='you')
df['word_freq_credit'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='credit')
df['word_freq_your'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='your')
df['word_freq_font'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='font')
df['word_freq_000'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='000')
df['word_freq_money'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='money')
df['word_freq_hp'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='hp')
df['word_freq_hpl'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='hpl')
df['word_freq_george'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='george')
df['word_freq_650'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='650')
df['word_freq_lab'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='lab')
df['word_freq_labs'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='labs')
df['word_freq_telnet'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='telnet')
df['word_freq_857'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='857')
df['word_freq_data'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='data')
df['word_freq_415'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='415')
df['word_freq_85'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='85')
df['word_freq_technology'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='technology')
df['word_freq_1999'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='1999')
df['word_freq_parts'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='parts')
df['word_freq_pm'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='pm')
df['word_freq_direct'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='direct')
df['word_freq_cs'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='cs')
df['word_freq_meeting'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='meeting')
df['word_freq_original'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='original')
df['word_freq_project'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='project')
df['word_freq_re'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='re')
df['word_freq_edu'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='edu')
df['word_freq_table'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='table')
df['word_freq_conference'] = df['Message'].apply(calcular_frecuencia_palabra, palabra='conference')

df['char_freq_;'] = df['Message'].apply(calcular_frecuencia_caracter, caracter=';')
df['char_freq_('] = df['Message'].apply(calcular_frecuencia_caracter, caracter='(')
df['char_freq_['] = df['Message'].apply(calcular_frecuencia_caracter, caracter='[')
df['char_freq_!'] = df['Message'].apply(calcular_frecuencia_caracter, caracter='!')
df['char_freq_$'] = df['Message'].apply(calcular_frecuencia_caracter, caracter='$')
df['char_freq_#'] = df['Message'].apply(calcular_frecuencia_caracter, caracter='#')

df['capital_run_length_average'] = df['Message'].apply(longitud_promedio_de_secuencias)
df['capital_run_length_longest'] = df['Message'].apply(longitud_de_la_secuencia_mas_larga)
df['capital_run_length_total'] = df['Message'].apply(numero_total_de_letras_mayusculas)
```
